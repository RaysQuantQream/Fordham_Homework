```python
import pandas as pd
import numpy as np
```

```python
# B.1
ticker = pd.read_csv('INTC_2012-06-21_orderbook_5min.csv')
```

```python
ticker
```

```python
# B.2
ticker['MidPrice'] = (ticker['AskPrice'] + ticker['BidPrice']) / 2
```

```python
ticker['BidAskSpread'] = ticker['AskPrice'] - ticker['BidPrice']
```

```python
ticker
```

```python
# B.3
start_market_value = 2000 * ticker['AskPrice'][0]
print('start_market_value: ',start_market_value)
sales_proceed = 2000 * ticker['BidPrice'].iloc[-1]
print('sales proceed at the end of the day: ', sales_proceed)
pnl = sales_proceed - start_market_value
print('P&L: ', pnl)
```

```python
# B.4.1
ticker['bought_share'] = round(ticker['AskSize'] *0.01, -2)
ticker['bought_share_total'] = ticker['bought_share'].cumsum()
num_of_5min = int(ticker[ticker.bought_share_total >= 2000].head(1).index.tolist()[0]) + 1
print("It takes %d minutes for the order to be filled"% (num_of_5min * 5))
```

```python
ticker
```

```python
# B.4.2
ticker_traded = ticker[:num_of_5min - 1] # in case the total share is more than 2000
ticker_traded = ticker_traded.copy()
ticker_traded['cost'] = ticker.loc[:, 'AskPrice'] * ticker.loc[:, 'bought_share']
ave_price = (ticker_traded['cost'].sum() + 
             (2000 - ticker_traded['bought_share_total'].iloc[-1]) * ticker.loc[num_of_5min - 1, 'AskPrice']) / 2000
print(ave_price)
```

```python
# B.4.3
pnl = ticker.iloc[-1]['BidPrice'] * 2000 - ave_price * 2000
print(pnl)
```
